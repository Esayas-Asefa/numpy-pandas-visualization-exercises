```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
print(df)
```

```python
1. 'Create a column named passing_english that indicates whether each student has a passing grade in english.'
```

```python
passing_english = df.english >= 70

df[passing_english]
```

```python
'2. Sort the english grades by the passing_english column. How are duplicates handled?''
```

```python
df[passing_english].sort_values(by='english')
```

```python
''''3. Sort the english grades first by passing_english and then by student name. All the students that are failing 
english should be first, and within the students that are failing english they should be ordered alphabetically. 
The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)'''
```

```python
p_e = df[passing_english].sort_values(by='english')
p_e.sort_values(by='name')
```

```python
'''4. Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did 
in the last step.'''
```

```python
p_e = df[passing_english].sort_values(by='english')
p_e.sort_values(by='english', ascending=False)
```

```python
''''5. Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average 
of the math, english, and reading grades.'''
```

```python
overall_grade =  (df.math + df.english + df.reading)/3
o_g = round(overall_grade, 2)
o_g
```

```python
Create a column named mileage_difference this column should contain the difference between highway and 
city mileage for each car.
```

```python
mpg['mileage_difference'] = round(mpg.hwy - mpg.cty, 2)
mpg[['hwy', 'cty', 'mileage_difference']].head()
```

```python
Which car (or cars) has the highest mileage difference?
```

```python
mpg.describe()
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]
```

```python
Which compact class car has the lowest highway mileage? The best?
```

```python
mpg[mpg['class'] == 'compact']
mpg[(mpg['class'] == 'compact') & 
    (mpg['hwy'] == mpg[mpg['class'] == 'compact'].hwy.max())]
```

```python
Create a column named average_mileage that is the mean of the city and highway mileage.
```

```python
mpg['avg_mileage'] = (mpg.cty + mpg.hwy) /2
mpg.head()
```

```python
Which dodge car has the best average mileage? The worst?
```

```python
mpg.columns
dodges = mpg[mpg.manufacturer == 'dodge']
dodges.head()
```

```python
dodges[dodges.avg_mileage == dodges.avg_mileage.max()]
dodges[dodges.avg_mileage == dodges.avg_mileage.min()]
```
