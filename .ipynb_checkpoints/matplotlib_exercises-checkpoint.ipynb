```python
plt.figure(figsize=(8,8))
plt.ylim(-1,8)
plt.xlim(-1,8)


plt.plot(x1,y1, label='$âˆšx$')
plt.plot(x2,y2, label='$x^3$')
plt.plot(x3,y3, label='$2^x$')
plt.plot(x4,y4, label='$1/(x+1)$')

plt.legend(loc='upper right')
plt.suptitle('Number 4 Multi-Plot Figure')

plt.show()
```
